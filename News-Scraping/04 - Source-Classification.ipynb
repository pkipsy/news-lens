### Load Bias Information

```python
#import pandas
import pandas as pd 

# define csv read
def read_data(filename):
    # read in csv
    df = pd.read_csv(filename, encoding='utf-8')
    
    #limit df content to bias, text, headline, and source
    #df = df.loc[:, ['bias', 'text', 'headline', 'source', 'sentiment', 'magnitude']]
    
    return df

# load cleaned file with sentiment data
ratings = read_data('allsides-media-bias-ratings.csv')
ratings.head()
```

### Launch Scraper

Create a text file of org keywords.

```python
# create keyword list to query web
ratings_list = []
for r in ratings['News Source'].values:
  new_r = r.replace(' ', '+')
  ratings_list.append(new_r)
  
query_file = open('query.txt', 'w')
for item in ratings_list:
  query_file.write("%s\n" % item)
```

This query produces a sqlite3 database.

```python
!GoogleScraper --keyword-file query.txt --search-engine google
```
